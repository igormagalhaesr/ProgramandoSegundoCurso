<h3>Busca</h3>

Vamos pensar no problema de buscar um elemento em uma lista ordenada, como buscar o nome de uma pessoa em uma antiga lista telefônica de papel. 
<br />
<i>Vou deixar só os nomes na lista pra simplificar, mas a forma de uma lista telefônica seria usar a estrutura conhecida como dicionário.</i>

Você pode checar nome por nome, página por página dessa lista até que encontre o nome que você quer. Esse algoritmo seria algo como:

```python
#Aqui temos a lista telefônica dividida por "páginas"
phoneBook = [["Abilio", "Alberto", "Astrid", "Ana Carolina"], 
             ["Barbara", "Beatriz", "Bruna", "Bruno"], 
             ["Carla", "Carlos", "Cecilia", "Clara",], 
             ["Dario", "Debora", "Diana", "Duilia"]]

#Agora vamos definir a função que procura um certo nome nessa lista e nos diz a posição
def buscaTelefonica(nomeLista, nomeProcurado):
    '''Essa função recebe uma lista e um nome, então busca linearmente 
    por uma string em uma lista bidimensional e retorna a posição.'''
    for i in range(len(nomeLista)):
        for j in range(len(nomeLista[i])):
            
            if nomeLista[i][j] == nomeProcurado:
                out = "O nome está na página " + str(i + 1) + " e na posição " + str(j + 1)
                
    return out

print(buscaTelefonica(phoneBook, "Ana Carolina"))
```

E essa é uma forma de resolver esse problema, mas você já pensou quanto tempo levaria se procurar cada nome levasse 1 segundo? Nesse caso teríamos 4 segundos, no pior caso, que é o nome estar na última posição da lista, teríamos 16 segundos pra percorrer essa lista inteira. 
<br />
<br />
Pense o que aconteceria se essa lista tivesse 8 milhões de nomes, o nosso pior caso levaria 8 milhões de segundos - que dá mais de três meses - pra ser achado. Ninguém tem todo esse tempo.

Vamos tentar resolver esse problema de uma forma melhor, tudo que vamos fazer é dividir de uma forma mais eficiente. Nós sabemos que essa lista está ordenada, então podemos começar procurando na metade da lista, se o nome começar com uma letra que fica antes de onde estamos procurando, nós nem precisamos mais se importar em procurar nessa metade depois da letra.

```python
print(int(len(phoneBook)/2))
print(phoneBook[2][0] == "Ana Carolina")
print(phoneBook[2][0])
```

Agora que nós sabemos que esse nome não está na segunda metade, podemos reduzir o problema pela metade. Agora vamos procurar só em 8 elementos em vez de procurar em 16. Mas e se nós continuarmos reduzindo isso pela metade?

```python
phoneBook = phoneBook[0:2]
print(int(len(phoneBook)/2))
```

Agora vamos procurar na metade dessa lista que restou

```python
print(phoneBook[1][0] == "Ana Carolina")
print(phoneBook[1][0])
```

Sabemos que o nome Barbara vem depois do nome Ana Carolina, então podemos ignorar a segunda metade desse problema de novo.  

```python
phoneBook = phoneBook[0:1]
```

Agora temos só uma página, mas temos quatro elementos nessa página. Vamos checar se o nome está na metade dessa página:

```python
print(phoneBook[0][2] == "Ana Carolina")
print(phoneBook[0][2])
```

Novamente não achamos o nome, mas podemos continuar dividindo a lista

```python
phoneBook = phoneBook[0][2:]
```

```python
print(phoneBook[1] == "Ana Carolina")
print(phoneBook[1])
```

E assim nós achamos o nome que queríamos achar. Percebeu alguma coisa? Com esse método nós precisamos de 4 passos pra achar o nome, ou seja, demoraríamos 4 segundos caso cada nome olhado levasse um segundo. A melhor parte desse método, no entanto, é que, como dividimos o problema pela metade a cada tentativa, no pior caso na lista de tamanho 16 nós levamos só 4 segundos. "E em uma lista com 8 milhões, quanto tempo demoraríamos?" você pode perguntar. Com esse método levaríamos apenas 22 segundos, comparados com os mais de três meses do outro.

Fizemos esse método - a busca binária - da forma manual, mas posteriormente iremos fazer uma abstração e implementar uma função que consiga fazer isso.

<h3>Abstração</h3>

A linguagem central da Ciência da Computação é a matemática, portanto - algo que diriam os logicistas - a linguagem central da Ciência da Computação é a lógica. Um conceito muito importante é a abstração, ou como podemos omitir certos detalhes pra focar no problema principal

Se quisermos, por exemplo, fazer uma combinação de m elementos tomados p a p, ou, como a fórmula diz:

Nós podemos primeiro definir uma função que faça o fatorial, 
<br />lembrando que n! = n * (n - 1) * (n - 2) * ... * 2 * 1 com n sendo um número natural maior que 5 nesse caso:

```python
def factorial(n):
    '''Essa função calcula o enésimo fatorial da forma recursiva'''
    if n == 0 or n == 1:
        out = 1
    
    elif n < 0 or int(n) != n:
        out = "erro"
    
    else:
        out = n * factorial(n - 1)
        
    return out
```

E agora nós podemos definir a função que faz a combinação sem ter que se preocupar com como o fatorial é feito, assim o fatorial funciona como uma caixa preta que recebe um n e retorna um n! Não precisamos saber o que acontece dentro da caixa, só precisamos do resultado.

```python
def combination(m, p):
    '''Essa função calcula a combinação de m elementos tomados p a p'''
    
    numerator = factorial(m)
    denominator = factorial(p) * factorial(m - p)
    
    out = int(numerator/denominator)
    
    return out
```

E, olhando só essa função, nós nem precisamos nos importar com como foi feito o fatorial, assim como usando a função combination nós não precisamos nos importar com a forma que ela foi implementada, e isso é abstração.
<br />
Quando fazemos <code>combination(5, 3)</code>, só precisamos saber a entrada - (5,3) - e a saída, o 10.

```python
combination(5,3)
```

<h2>Paradigmas de Programação</h2>

<b>Programação Orientada a Objetos</b> é um <i>Paradigma de Programação</i>, que é basicamente um estilo de programação, com sua estrutura própria, suas formas de se fazer as coisas e suas vantagens. 

Pra aprender um novo paradigma, no entanto, precisamos entender como nós estávamos programando antes. <h2>Programação Imperativa</h2>


A Programação Imperativa é um paradigma que talvez seja o mais simples, é baseado em dar comandos no programa, passo a passo, com o programa seguindo exatamente a ordem desses comandos. O programa usa esses comandos em ordem pra alterar seus estados.

```python
paresAteDez = []

for i in range(10):
    if i % 2 == 0:
        paresAteDez += [i]
        
print(paresAteDez)
```

<h2>Programação Procedural</h2>

A programação procedural é um lugar no qual usamos muito a abstração. Uma <i>subrotina</i>, ou como chamamos, uma <b>função</b>, é uma forma de evitar repetir um bloco de código várias vezes ao longo do seu programa. Usando esse recurso, pode-se definir uma função uma só vez, e usá-la quando quiser, com as variáveis sendo locais da função.

```python
def paresAte(n):
    out = []
    for i in range(n):
        if i % 2 == 0:
            out += [i]
            
    return out
            
print(paresAte(10))
```

<h2>Programação Orientada a Objetos</h2>

Nesse paradigma, como o nome diz, nós pensamos nas coisas como objetos. Um objeto tem certos atributos (dados) e certas funções que são só dele, nós chamamos essas funções de métodos.

```python
class ListaDePares(object): #aqui definimos a classe ListaDePares, poderemos instanciar objetos dessa classe
    '''Um conjunto de números pares'''
    
    def __init__(self): #o método __init__ é um construtor, cria o objeto e prepara pra usar 
        #self representa a instância do objeto
        '''Cria uma lista vazia de números pares'''
        self.vals = []
        
    def inserir(self, n):
        '''Assume que n é um número par e insere em self'''
        if n not in self.vals:
            self.vals.append(n)
            
    def membros(self):
        '''Retorna uma lista com os elementos de self'''
        return self.vals[:]
        
```

```python
listaPares = ListaDePares()
for i in range(0, 10, 2):
    listaPares.inserir(i)
    
print(listaPares.membros())
```

Você consegue perceber o quanto o código ficou mais abstrato? No primeiro caso nós temos algo que funciona ali no momento, no segundo nós já definimos uma função pro código inteiro, no terceiro fizemos uma classe de objetos que têm seus próprios dados e métodos. Esse  tipo de abstração pode parecer à toa em um código são simples, mas é muito útil na medida que o nosso programa vai ficando maior e mais complexo.

Saber o que são paradigmas e conhecer alguns tipos é importante pra saber quando é mais vantajoso programar de certa forma. O Python é uma linguagem orientada a objetos, o que significa que os dados que vemos são objetos, e têm seus próprios métodos e características. 

Pra adicionar um dado em uma certa posição numa lista, por exemplo, em vez de fazer:

```python
lista = [1, 2, 3, 4, 6]
listaVazia = []

for i in range(len(lista) + 1):
    listaVazia += [i + 1]
    if lista[i] == 4:
        lista += [5]
        
lista = listaVazia

print(lista)
```

Podemos simplesmente fazer

```python
lista = [1, 2, 3, 4, 6]
lista.insert(4, 5)
print(lista)
```

O método pra inserir um elemento em uma certa posição de uma lista já foi implementado no python, assim como vários outros métodos. Conhecendo a documentação, em vez de ter todo o trabalho pra pensar em um algoritmo que ordena uma lista, como:

```python
lista = [2, 6, 1, 3, 5, 4]

for i in range(len(lista)):
    for j in range(len(lista)-1):
        if lista[j] > lista[j + 1]:
            lista[j], lista[j + 1] = lista [j + 1], lista[j]
            
print(lista)
```

<h2>1b. Herança</h2>

Herança é uma forma de criar ainda mais abstrações fazendo classes dentro de classes, assim a <b>subclasse</b> - que está dentro de outra - herda os atributos da <b>superclasse</b>, que a contém.

Quando pensamos na classe cachorro, por exemplo, também podemos definir a classe labrador:

```python
class Cachorro(object):
    def __init__(self, nome):
        self.nome = nome
        
    def latir(self):
        print(self.nome.title() + " está latindo")
    
    def comer(self):
        print(self.nome.title() + " está comendo")
```

```python
class Labrador(Cachorro):
    def __init__(self, nome):
        Cachorro.__init__(self, nome) 
        
    def latirEmLabrador(self):
        print(self.nome.title() + " está latindo em labradorês")
```

Assim nós podemos definir o objeto rex, da classe labrador, que tem a propriedade de saber latir labradorês

```python
rex = Labrador("rex") #essa linha cria um novo Labrador chamado rex

rex.latirEmLabrador() #aqui nós usamos um método associado a Labrador
```

Mas como sabemos que Rex é um cachorro e cachorros também podem comer, rex herda essa propriedade:

```python
rex.comer() #nessa linha é testado se comer é um método de Labrador
            #como não é o caso, testa-se depois se comer é um método da superclasse Cachorro
```

Podemos ver, então, a representação de superclasses e subclasses como uma árvore, seus galhos mais finos herdam os atributos dos galhos mais grossos. 
<br />
A melhor forma de pensar na hierarquia é como subclasses sendo uma forma de extender o comportamento das suas superclasses. Criar métodos que anulam um efeito da classe mais específica é algo que deve ser feito com cuidado.

Tanto labrador quanto beagle têm seus atributos próprios, mas ambos herdam os atributos do cachorro. 

Se um mesmo atributo for definido na subclasse e na superclasse, o atributo da subclasse sempre prevalece, assim, se definirmos que cachorro tem pelo grande, mas que labrador não tem pelo grande, não ter o pelo grande vai prevalecer.

Se o pelo grande não é um atributo geral de cachorro, no entanto, talvez nem devêssemos ter definido isso.

<h2>1c. Encapsulamento</h2>

Encapsulamento é o processo de esconder os atributos e como funcionam os métodos da nossa classe. Quando fazemos uma classe, precisamos ter certeza que seus dados não podem ser alterados por outra parte do nosso programa.

Nosso labrador, o rex, pode ter seu nome mudado de qualquer lugar do código, isso nos gera um problema com privacidade. Ainda, isso possivelmente vai gerar bugs, já que um programador alterando a implementação da classe poderia quebrar esse código, que é acessado de fora.

```python
rex.nome = "spark"

print(rex.nome)
```

Pra resolver isso nós podemos deixar <b>privados</b> certos atributos da nossa classe, só permitindo que sejam acessados e alterados por meio de métodos que deixamos públicos.

Vamos pensar no seguinte problema: temos um cachorro que está fora do peso, queremos criar uma classe pra acompanhar a rotina dele ao emagrecimento.

```python
class CachorroPesado(object):
    def __init__(self, nome, peso):
        self.nome = nome
        self.peso = peso
        
    def pesar(self, peso):
        print(self.peso)
```

```python
rex = CachorroPesado("rex", 40)
```

Assim, se o rex perdesse peso nós poderíamos simplesmente alterar diretamente o atributo peso:

```python
rex.peso = 35
rex.pesar(rex.peso)
```

No entanto, alguém que quisesse dar petiscos ao rex poderia simplesmente mudar isso pro valor que quiser, quando quiser. Podemos deixar o atributo peso privado, só deixando mudar com os métodos que definimos. Como fazemos isso no python, no entanto?

É simples, devemos usar a seguinte notação:

```python
class CachorroPesado(object):
    def __init__(self, nome, peso):
        self._nome = nome
        self.__peso = peso
        
    def pesar(self, peso):
        print(self.__peso)
```

Quando definimos um atributo com dois <i>underscores</i> antes do nome, estamos falando pro Python que esse dado é privado, ou seja, só pode ser alterado dentro da própria classe. É o que fizemos com <code>self.__peso</code>

Quando definimos um atributo com apenas um <i>underscore</i> antes do nome, estamos seguindo uma convenção que foi decidida no python. Estamos falando pra quem lê que aquela variável não deve ser diretamente acessada.

```python
rex = CachorroPesado("rex", 40)
```

```python
print(rex._nome)
```

```python
print(rex.__peso)
```

Assim o peso não pode ser alterado por fora da classe. Nesse exemplo, no entanto, nem definimos métodos que possibilitem isso. Vamos fazer agora:

```python
class CachorroPesado(object):
    def __init__(self, nome, peso):
        self._nome = nome
        self.__peso = peso
        
    def getPeso(self):
        return self.__peso
    
    def setPeso(self, peso):
        self.__peso = peso
```

```python
rex = CachorroPesado("rex", 40)
```

```python
rex.setPeso(35)
print(rex.getPeso())
```

Os métodos <b>get</b> e <b>set</b>, ou <i>acessar</i> e <i>definir</i> garantem que encapsulemos nossos dados. São a convenção a se usar nas linguagens orientadas a objeto. No python, no entanto, nós costumamos não usar essa forma de esconder os dados - como não vamos usar nesse livro.

A forma pythônica de se programar é priorizando a legibilidade do código, e esses <i>underscores</i> podem atrapalhar isso. No python, inclusive, o seguinte código funciona sem problemas: 

```python
rex.peso = 20
```

Nós encapsulamos o dado, mas quando atribuímos valor a um atributo que não existe, o python - por ser uma linguagem dinâmica - simplesmente cria esse novo atributo no nosso objeto. Assim, temos:

```python
print(rex.getPeso())
print(rex.peso)
```

Por isso o melhor a se fazer quando programamos em python é se disciplinar e seguir a regra de não acessar atributos de fora da classe, é isso que faremos aqui.

<h2>2a. Eficiência</h2>

Nós definimos, então, uma proprieadade chamada <b>eficiência</b>, que mede justamente a quantidade de recursos computacionais usados por esse algoritmo.<br />
<br />
Como a gente pode, no entanto, medir o quão bom um certo algoritmo é comparado a outro? Você pode pensar no tempo que ele leva pra rodar, mas seguindo essa estratégia vamos nos deprar com alguns problemas: <br />
<br />
Isso depende da máquina que você está rodando o programa, da velocidade da linguagem que você está usando e do valor do input. <br />
<br />
Vamos ver um exemplo, testar se um número é primo:

```python
def primalityTest(n):
    '''Testa se um número inteiro é primo'''
    
    prime = True
    for i in range(2, n):
        if n % i == 0:
            prime = False
            
    return prime
```

Com esse algoritmo, nós sempre vamos testar se o número é primo <b>n</b> vezes. Assim, checando se o número 23 é primo, testaremos 23 vezes. <br />
<br />
Nós podemos fazer uma pequena mudança e melhorar muito a eficiência desse algoritmo.

```python
def primalityTest2(n):
    '''Testa se um número inteiro é primo'''
    
    limit = int(n ** (1/2))
    prime = True
    for i in range(2, limit):
        if n % i == 0:
            prime = False
            
    return prime
```

Por que nós podemos fazer isso? Simples, se n é divisível por um número p, então n = p x q (ele é o produto de dois números, com p sendo um deles) e se q fosse menor que p, então q seria detectado como um divisor de n antes.

No caso  de 24 = 8 * 3, por exemplo, quando testamos se 24 é divisível por 8 nós já tínhamos testado se era divisível por 3, já que vamos testando em ordem crescente a partir do 2. No pior caso nós devemos testar, então, a raíz do número (arredondada).

Testando se o número 10 é primo das duas formas:

```python
print(primalityTest(100))
print(primalityTest2(100))
```

Nós conseguimos o mesmo resultado das duas formas, mas no primeiro nós testamos 100 vezes, enquanto precisamos testar apenas 10 no segundo.

Podemos usar o módulo <code>time</code> do python pra comparar o tempo entre esses dois, só por curiosidade.

```python
import time 

def primalityTest(n):
    '''Testa se um número inteiro é primo'''
    
    start = time.time()
    prime = True
    for i in range(2, n):
        if n % i == 0:
            prime = False
        
    end = time.time()
    
    return prime, end-start

def primalityTest2(n):
    '''Testa se um número inteiro é primo'''
    
    start = time.time()
    limit = int(n ** (1/2))
    prime = True
    for i in range(2, limit):
        if n % i == 0:
            prime = False
    
    end = time.time()
            
    return prime, end-start
```

Lembramos que o tempo de execução é uma medida que varia muito, de computador pra computador, de linguagem pra linguagem. Precisamos de uma forma de comparar algoritmos que seja mais geral, mais abstrata.

<h2>2b. Notação Assintótica</h2>

A maneira que foi achada pra comparar os algoritmos de forma mais abstrata é caracterizá-los em termos de passos que ele executa. <br />
<br />
Considerando cada passo como uma unidade, podemos comparar algoritmos de acordo com a quantidade desses passos que cada um requer pra executar uma certa tarefa.

Assim, no exemplo que usamos, podemos ver que o seguinte programa:

```python
def primalityTest(n):
    '''Testa se um número inteiro é primo'''
    
    prime = True
    for i in range(2, n):
        if n % i == 0:
            prime = False
            
    return prime
```

Vai executar <code>n</code> passos no loop até terminar, enquanto o segundo programa:

```python
def primalityTest2(n):
    '''Testa se um número inteiro é primo'''
    
    limit = int(n ** (1/2))
    prime = True
    for i in range(2, limit):
        if n % i == 0:
            prime = False
            
    return prime
```

vai executar <code>&#8730;n</code> passos no loop antes de resolver o problema.

Por que, no entanto, nós não contamos os outros passos? No programa também tivemos, por exemplo <code>prime = True</code>, mesmo assim estamos contando só os n passos.<br />
<br />
A verdade é que, na medida que o <code>n</code> cresce, esses outros passos pouco afetam a execução do nosso código, então nós fazemos essa técnica de análise de algoritmos en função do termo dominante, que é quem domina o problema na medida que ele cresce.

Pra falar dessa <i>complexidade</i> ou <i>ordem de magnitude</i> do problema, usamos a notação <b>Big-O</b>. Assim, o <code>primalityTest</code> tem complexidade <b>O(n)</b>, enquanto o <code>primalityTest2</code> tem complexidade O(&#8730;n).


<center><i>Em azul temos n, em vermelho temos &#8730;n.</i></center>

Podemos ver que, pra qualquer valor de n maior que 1, o tempo de execução - representado por y - aumenta de forma muito mais rápida em <b>n</b> que em <b>&#8730;n</b>. <br />
<br />
Quando falarmos na complexidade aqui, estaremos falando do pior caso possível, mas também podemos analisar o melhor caso ou o caso médio de um algoritmo.

Vamos pensar em um algoritmo que busca linearmente em uma lista.

```python
def buscaTelefonica(nomeLista, nomeProcurado):
    '''Essa função recebe uma lista e um nome, então busca linearmente 
    por uma string em uma lista e retorna a posição.'''
    for i in range(len(nomeLista)):
        if nomeLista[i] == nomeProcurado:
            out = "O nome está na posição " + str(i + 1)
                
    return out
```

Temos no pior caso, que é quando o elemento está na última posição da lista, n comparações, então o pior caso tem complexidade O(n). No melhor caso temos O(1), que seria com o elemento procurado estando na última posição.<br />
<br />

Pra calcular o caso médio, devemos somar os casos de todos os inputs possíveis e dividir pela soma do número de inputs. No caso da busca linear o caso médio seria O(n/2).<br />
<br />

Quando estamos pensando na complexidade, no entanto, pensamos no que acontece quando o n tende a um número muito grande, quando tende ao infinito. Nesse caso não teria diferença entre n e n/2, então podemos dizer que o caso médio da busca linear é O(n).

<i><h3>Constante - O(1)</h3></i>

O tempo constante, ou O(1), acontece quando a complexidade do algoritmo não depende do tamanho do input.



Um exemplo de um código que roda em tempo constante:

```python
def compareAnB(a, b):
    start = time.time()
    if a > b:
        out = True
    else:
        out = False
    end = time.time()
    
    return "O programa levou " + str(end - start) + " segundos"
```

```python
print(compareAnB(2 ,3))
print(compareAnB(12334254, 923787))
```

Como é possível ver, o tempo que esse programa leva pra rodar não depende do tamanho de a ou de b.<br />
<br />
O seguinte código:

```python
def productFiveSum(a):
    '''Essa função define o tempo do produto de um número a por 5, usando apenas soma'''
    start = time.time()
    inicial = a
    for i in range(4):
        a += inicial
        
    end = time.time()
         
    return "O programa levou " + str(end - start) + " segundos"
```

```python
print(productFiveSum(5))
print(productFiveSum(3246557))
```

Pode-se ver que o programa vai rodar em um tempo constante independentemente do tamanho de <code>a</code>.

<i><h3>Linear - O(n)</h3></i>

Um algoritmo é dito ter tempo linear, ou complexidade O(n), quando o maior tempo possível pra ele rodar cresce linearmente com o tamanho do seu input. 



Um exemplo de algoritmo que tem um tempo de execução que cresce linearmente de acordo com o input é o de achar o fatorial da seguinte forma:

```python
def factorial(x):
    start = time.time()
    if x == 0 or x == 1:
        out = 1
    else:
        out = x * factorial(x - 1)
    end = time.time()
    
    return "O programa levou " + str(end - start) + " segundos"
```

```python
print(factorial(20))
print(factorial(30))
print(factorial(40))
print(factorial(50))
print(factorial(60))
print(factorial(70))
```

O tempo de execução do programa aumenta no máximo linearmente na medida que o input x fica maior.

<i><h3>Logarítmica - O(log(n))</h3></i>

Esse tipo de algoritmo tem uma complexidade que cresce como o logaritmo de um dos inputs. 



Um exemplo seria a busca binária, que será explicada de forma mais detalhada posteriormente:

```python
def binarySearch(array, left, right, element):
    '''Busca por um elemento em uma lista ordenada'''
    
    mid = (left + right) // 2
    
    if array[mid] == element:
        out = mid
        
    elif array[mid] > element:
        out = binarySearch(array, left, mid - 1, element)

    else:
        out = binarySearch(array, mid + 1, right, element)
        
    return out
```

```python
arr = [1, 5, 7, 21, 45, 51]
left = 0
right = len(arr) - 1

print(binarySearch(arr, left, right, 21))
```

Alguns algoritmos como o merge sort e o quicksort aleatorizado têm essa complexidade, mas veremos mais sobre eles à frente.



<i><h3>Quadrática - O(n²)</h3></i>

Nesse tipo, o tempo de execução do algoritmo aumenta com o quadrado do input.



Um algoritmo que tem complexidade O(n²) é o bubble sort:

```python
def bubbleSort(array):
    
    for i in range(len(array)):
        for j in range(len(array) - 1):
            if array[j] > array[j + 1]:
                    array[j], array[j + 1] = array[j + 1], array[j]
                    
    return array
```

Algoritmos que o tempo de execução cresce no máximo com Cn<sup>k</sup> com C e k sendo constantes positivas, como O(n²) e O(n³), são algoritmos que rodam em <b>tempo polinomial</b>. <br />
<br />
Problemas que têm um algoritmo determinístico polinomial são da classe de complexidade <b>P</b>, em teoria da complexidade computacional.
