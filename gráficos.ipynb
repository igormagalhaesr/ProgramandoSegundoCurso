```python
import matplotlib
from sympy import *
import numpy
from matplotlib import pyplot
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
print("Comparação entre linear e raiz de n")
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

x = numpy.linspace(1, 100, 100);
y = numpy.linspace(1, 10, 10)
axes = pyplot.gca()
axes.set_xlim([1,100])
axes.set_ylim([1,10])

b = x
a = numpy.sqrt(x)

pyplot.plot(x, a, 'r')
pyplot.plot(x, b, 'b')

pyplot.grid()
pyplot.show()
```

```python
print("Comparação")
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

x = numpy.linspace(1, 100, 1000);
y = numpy.linspace(1, 10, 10)
axes = pyplot.gca()
axes.set_xlim([1,10])
axes.set_ylim([1,10])
logx = numpy.log(x)

logarithmic = numpy.log2(x)
linear = x
logLinear = x * logx
quadratic = x ** 2
cubic = x ** 3
exponential = numpy.exp2(x)

pyplot.plot(x, cubic, 'm', label='cúbica')
pyplot.plot(x, quadratic, 'c', label='quadrática')
pyplot.plot(x, exponential, 'y', label='exponencial')
pyplot.plot(x, logLinear, 'b', label='log-linear')
pyplot.plot(x, linear, 'g', label='linear')
pyplot.plot(x, logarithmic, 'r', label='logarítmico')

pyplot.legend()
pyplot.grid()
pyplot.show()
```

```python
print("Comparação log e linear")
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

x = numpy.linspace(1, 100, 1000);
y = numpy.linspace(1, 10, 10)
axes = pyplot.gca()
axes.set_xlim([1,10])
axes.set_ylim([1,10])
logx = numpy.log(x)

logarithmic = numpy.log2(x)
linear = x

pyplot.plot(x, linear, 'b', label='linear')
pyplot.plot(x, logarithmic, 'r', label='logarítmico')

pyplot.legend()
pyplot.grid()
pyplot.show()
```
